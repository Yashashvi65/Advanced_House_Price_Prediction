## Missing Values

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing values')
```

#### Finding the relation bw the missing values and the sales price

```python
for feature in features_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### We see that the missing values and the dependent variable are clearly related so we need to replace these missing values with something meaningful which we will see in feature engineering

## Numerical Variables

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtype !='O']
print("Number of numerical variables:",len(numerical_features))
dataset[numerical_features].head()
```

### Temporal Variables need to be handled

```python
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.title('House Price vs YearSold')
```

```python
year_feature
```

```python
#Compare diff bw all years feature with SalePrice
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        #diff bw year var and the year the house was sold for
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```
